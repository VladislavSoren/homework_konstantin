```python
class MyRange():
    def __init__(self, n):
        self.n = n
        self.current = 0

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current < self.n:
            value = self.current
            self.current += 1
            return value
        else:
            raise StopIteration
    
```

```python
custom_range = MyRange(3)
iterator = iter(custom_range)
next(iterator)
next(iterator)
next(iterator)
```
