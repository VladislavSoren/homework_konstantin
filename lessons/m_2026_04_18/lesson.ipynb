```python
# атрибуты объекта класса
class User:
    is_human = True  # атрибут класса is_human
    name = "Standart Name"  # атрибут класса name

    # def __init__(self, age): pass

    def __init__(self, age):
        self.age = age  # атрибуты объекта класса
#        super().__init__()

    # метод класса
    @classmethod
    def print_class_info(cls): #  cls - ссылка на класс User
        print(f"is_human - {cls.is_human}, name - {cls.name}, age - {cls.age}")

    # метод объекта класса
    def print_object_info(self): #  self - ссылка на объект класса User, т.е. на user (user = User())
        print(f"is_human - {self.is_human}, name - {self.name}, age - {self.age}")
```

```python
# мы не можем из класса обратиться к методу объекта класса  
User.print_class_info()
```

```python
user = User(23)
user.print_object_info()
```

```python
User.print_object_info(user)
```

```python
# атрибуты объекта класса
class User:
    is_human = True  # атрибут класса is_human
    name = "Standart Name"  # атрибут класса name

    # конструктор класса, через него мы можем наполнять объект новыми свойствами
    def __init__(self, age):
        self.age = age  # атрибуты объекта класса

    # метод класса
    @classmethod
    def print_class_info(cls): #  cls - ссылка на класс User
        print(f"is_human - {cls.is_human}, name - {cls.name}, age - {cls.age}")

    # метод объекта класса
    def print_object_info(self): #  self - ссылка на объект класса User, т.е. на user (user = User())
        print(f"is_human - {self.is_human}, name - {self.name}, age - {self.age}")
```

```python
# мы не можем из класса обратиться к атрибуту объекта класса  
print(User.is_human)
print(User.name)
print(User.age)
```

```python
class Person:
    def __new__(cls, *args, **kwargs):
        print("1. Вызван __new__: Создание объекта")
        return super().__new__(cls)

    def __init__(self, name):
        print("2. Вызван __init__: Инициализация объекта")
        self.name = name

# Создание объекта
p = Person("Ivan")
p
```

```python
class Person:
    def __new__(cls, *args, **kwargs):
        print("1. Вызван __new__: Создание объекта")
        return super().__new__(cls)

    def __init__(self, name):
        print("2. Вызван __init__: Инициализация объекта")
        self.name = name

# Создание объекта
p = Person("Ivan")
p
```
